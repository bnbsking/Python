```python
import random
```

```python
def binarySearchInsert(A,target):
    l, r = 0, len(A)-1
    while l<=r:
        m = (l+r)//2
        if A[m]==target:
            return m
        elif A[m]>target:
            r = m-1
        else:
            l = m+1
    return l

for _ in range(100):
    A = sorted([ random.randint(1,100) for __ in range(100) ])
    x = random.randint(-10,110)
    
    l = binarySearchInsert(A,x)
    A.insert(l,x)
    assert A==sorted(A), (A, sorted(A))
```

```python
def quickSort(L):
    def f(l=0, r=len(L)-1):
        #print(l, r, L, L[l:r+1])
        if r-l<=1: # len<=2
            L[l], L[r] = min(L[l],L[r]), max(L[l],L[r])
        else:
            pivot, i, j = L[l], l+1, r
            while i<j:
                while i<j and L[i]<=pivot:
                    i+=1
                while i<j and L[j]>=pivot:
                    j-=1
                L[i], L[j] = L[j], L[i]
            # from now on, i==j and L[idx<i] <= pivot <= L[idx>i], but pivot ? L[i] # e.g. 5 1 2 L[i] 6 7
            i -= int(pivot<L[i]) # must go back 1 # SUPER IMPORTNANT
            L[l], L[i] = L[i], L[l]
            # from now on, L[idx<=i] <= pivot <= L[idx>i]
            
            #print("*", l, r, i, L, "\n"+"-"*10)
            f(l,i-1) if l<i-1 else None # prevent len=1,0,-1
            f(i+1,r) if i+1<r else None # prevent len=1,0,-1
    f()

for _ in range(100):
    A = [ random.randint(1,100) for __ in range(100) ]
    
    B = A[:]
    quickSort(B)
    assert sorted(A)==B, (sorted(A), B)
```

```python
def findIthSmallest(ith,L): # O(n)
    ith-=1
    def find(l=0, r=len(L)-1): # inclusive, ith is constant
        #print(l, r, L, L[l:r+1])
        if r-l<=1: # len<=2
            L[l], L[r] = min(L[l],L[r]), max(L[l],L[r])
            return L[ith]
        else:
            pivot, i, j = L[l], l+1, r
            while i<j:
                while i<j and L[i]<=pivot:
                    i+=1
                while i<j and L[j]>=pivot:
                    j-=1
                L[i], L[j] = L[j], L[i]
            # from now on, i==j and L[idx<i] <= pivot <= L[idx>i], but pivot ? L[i] # e.g. 5 1 2 L[i] 6 7
            i -= int(pivot<L[i]) # must go back 1 # SUPER IMPORTNANT
            L[l], L[i] = L[i], L[l]
            # from now on, L[idx<=i] <= pivot <= L[idx>i]

            #print("*", i, L, "\n"+"-"*10)
            if ith==i:
                return L[i]
            elif ith<i:
                return find(l,i-1)
            else:
                return find(i+1,r)
    return find()

for _ in range(100):
    L = [ random.randint(1,100) for __ in range(100) ]
    ith = random.randint(1,100)
    
    assert findIthSmallest(ith, L) == sorted(L)[ith-1], L
```

```python
def mergeSort(L):
    if len(L)<2:
        return L
    else:
        leftSorted = mergeSort(L[:len(L)//2])
        rightSorted = mergeSort(L[len(L)//2:])
        merge = []
        l, r = 0, 0 
        while True:
            if l==len(leftSorted):
                merge+=rightSorted[r:]
                break
            elif r==len(rightSorted):
                merge+=leftSorted[l:]
                break
            else:
                if leftSorted[l]<=rightSorted[r]:
                    merge.append( leftSorted[l] )
                    l+=1
                else:
                    merge.append( rightSorted[r] )
                    r+=1
        return merge
    
for _ in range(100):
    L = [ random.randint(1,100) for __ in range(100) ]
    
    assert mergeSort(L) == sorted(L), L
```

```python
def heapSort(L):
    import heapq
    heapq.heapify(L)
    return [ heapq.heappop(L) for _ in range(len(L)) ]
    
for _ in range(100):
    L = [ random.randint(1,100) for __ in range(100) ]
    
    assert heapSort(L[:])==sorted(L), L
```
