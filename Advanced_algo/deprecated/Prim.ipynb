```python
class Solution(object):
    def minCostConnectPoints(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """
        G = { (x,y):{} for x,y in points }
        for i in range(len(points)-1):
            for j in range(i+1,len(points)):
                (lx, ly), (rx, ry)  = points[i], points[j] 
                G[(lx,ly)][(rx,ry)] = abs(lx-rx) + abs(ly-ry)
                G[(rx,ry)][(lx,ly)] = abs(lx-rx) + abs(ly-ry)
        
        start = tuple(points[0])
        S = { (G[start][nb],start,nb) for nb in G[start] }
        T, inS = { t:{} for t in G }, {start}
        #print( "S=", S, "T=", T, "inS=", inS )
        for i in range(len(G)-1):
            minw, minl, minr = min(S)
            #print(min(S))
            T[minl][minr], T[minr][minl] = minw, minw
            inS.add( minr )
            S = S.union( { (G[minr][nb],minr,nb) for nb in G[minr] } )
            S = { (w,l,r) for w,l,r in S if (r not in inS ) }
            #print( "S=", S, "T=", T, "inS=", inS ) # and (w,l,r)!=(minw,minr,minl)
            #print( "-"*10 )
            print( len(S) )
            
        ans = 0
        for k1 in T:
            for k2 in T[k1]:
                ans += T[k1][k2]
        return ans/2
    
Solution().minCostConnectPoints(W)
```
